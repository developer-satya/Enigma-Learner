```python
# import everything from tkinter module
from tkinter import *   
import os
import pywhatkit
 
# create a tkinter window
root = Tk()              
 
# Open window having dimension 100x100
root.geometry('400x500') 
 
# Create a Button
btn1 = Button(root, text = "Camera", bd = '5',command = os.system("notepad")) 
# Create a Button
btn2 = Button(root, text = "email", bd = '5',
                          command = Mailing()) 
# Create a Button
btn3 = Button(root, text = 'google search', bd = '5',
                          command = searchQuery())

btn3 = Button(root, text = 'whatsapp', bd = '5',
                          command = sendWhatmsg())
 
# Set the position of button on the top of window.   
btn1.pack(side = 'top')    
btn2.pack(side = 'top')  
btn3.pack(side = 'top') 


def sendWhatmsg():
    msg = input("Enter sms: ")
    pywhatkit.sendWhatmsg("7992160083","Hello Satya!", "16",'45')

def searchQuery():
    query = input("Enter query: ")
    pywhatkit.searchQuery(query)
    

def Mailing():
    receiver_email = input("Enter receiver e-mail: ")
    subject = input("Enter Subject: ")
    message = input("Enter e-mail body: ")
    sendEmail(receiver_email, subject, message)

def sendEmail(receiver_email, subject, email_body):
    smtp_server = 'smtp.gmail.com'
    smtp_port = 587  
    sender_email = os.environ['EMAIL']
    sender_password = os.environ['EMAIL_APP_PASSWORD']
    try:
        server = smtplib.SMTP(smtp_server, smtp_port)
        server.starttls()  # Start TLS encryption
        server.login(sender_email, sender_password)

        # Compose the email
        email_body = f'Subject: {subject}\n\n{message}'

        # Send the email
        server.sendmail(sender_email, receiver_email, email_body)

        # Close the SMTP server
        server.quit()

        print('Email sent successfully')
    except Exception as e:
        print(f'Email could not be sent. Error: {str(e)}')
        
root.mainloop()
```

```python
import cv2

# Open a connection to the camera (0 represents the default camera)
cap = cv2.VideoCapture(0)

# Set video width and height
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

overlay_image1 = cv2.imread('./doss.jpg')
# cv2.imshow("Image",overlay_image)
overlay_image = cv2.resize(overlay_image1,(100,100))

while True:
    # Read a frame from the camera
    ret, frame = cap.read()

    # Check if the frame is read successfully
    if not ret:
        print("Error: Failed to capture frame.")
        break
        
    frame[50:10, 50:10] = overlay_image

    # Display the frame
    cv2.imshow("Video Capture", frame)

    # Break the loop when 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera and close the window
cap.release()
cv2.destroyAllWindows()
```

```python
import tkinter as tk
import pandas as pd  # Import Pandas module

user_emails = []

def save_entries(name_entry):
    name = name_entry.get()
    user_emails.append(name)
    print(name)

def show_emails():
    # Create a DataFrame from the collected emails
    df = pd.DataFrame({'Emails': user_emails})
    
    # Display the DataFrame
    top = tk.Toplevel()
    top.title("Emails List")

    text_widget = tk.Text(top)
    text_widget.insert(tk.END, df)
    text_widget.pack()

def new_user():
    r = tk.Tk()
    r.title("Made for You")

    # Entry for email
    name_label = tk.Label(text='Email', bg='white', fg='red')
    name_label.pack()

    name_entry = tk.Entry(r)
    name_entry.pack()

    save_button = tk.Button(r, text='Save', width=10, command=lambda: save_entries(name_entry))
    save_button.pack()

    show_button = tk.Button(r, text='Show Emails', width=15, command=show_emails)
    show_button.pack()

    r.mainloop()

new_user()
print("Program RunÂ Successfully")
```
